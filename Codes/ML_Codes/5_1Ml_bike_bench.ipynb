```python
#Import dependencies

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
# Load the X_train, test; y_train, test files

from google.colab import drive
drive.mount('/content/drive')


X_train = pd.read_pickle('/content/drive/MyDrive/final/X_train.pkl')
X_test = pd.read_pickle('/content/drive/MyDrive/final/X_test.pkl')
y_train = pd.read_pickle('/content/drive/MyDrive/final/y_train.pkl')
y_test = pd.read_pickle('/content/drive/MyDrive/final/y_test.pkl')
```

## LinearRegression

```python
# train model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Evaluation
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error on test set: {mae:.2f} minutes")
```

```python

# MAE (already used)
mae = mean_absolute_error(y_test, y_pred)

# RMSE (Root Mean Squared Error)
rmse = mean_squared_error(y_test, y_pred)

# RÂ² Score (coefficient of determination)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f} minutes")
print(f"RMSE: {rmse:.2f} minutes")
print(f"RÂ² Score: {r2:.4f}")

#MAE shows average error in the same units.
#RMSE penalizes large errors more heavily.
#RÂ² Score shows how much variance in duration is explained by the model
#   (1.0 is perfect, 0.0 means "as good as mean", negative means worse).
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.xlabel("Actual Duration (min)")
plt.ylabel("Predicted Duration (min)")
plt.title("Actual vs Predicted Duration")
plt.plot([0, 120], [0, 120], 'r--', linewidth=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```
