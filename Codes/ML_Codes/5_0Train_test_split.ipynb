```python
#Import dependencies 

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Load the CSV file
df = pd.read_csv('archivo_ordenado.csv')

# Display the first 5 rows
df.head()
```

```python
# Convert date and time columns to datetime

df['Fecha_Hora_Retiro'] = pd.to_datetime(df['Fecha_Retiro'] + ' ' + df['Hora_Retiro'], dayfirst=True)
df['Fecha_Hora_Arribo'] = pd.to_datetime(df['Fecha_Arribo'] + ' ' + df['Hora_Arribo'], dayfirst=True)

# Calculate ride duration
df['Duracion_Viaje'] = (df['Fecha_Hora_Arribo'] - df['Fecha_Hora_Retiro']).dt.total_seconds() / 60
```

```python
# Drop rows with invalid duration
df = df[(df['Duracion_Viaje'] > 10) & (df['Duracion_Viaje'] < 120)]
```

```python
df['Hora_Retiro_dt'] = pd.to_datetime(df['Hora_Retiro'], format='%H:%M:%S', errors='coerce')

# Extract hour as integer
df['Hora_Dia'] = df['Hora_Retiro_dt'].dt.hour
```

```python
df['Fecha_Hora_Retiro'] = pd.to_datetime(df['Fecha_Retiro'] + ' ' + df['Hora_Retiro'], dayfirst=True)
```

```python
# Monday = 0, Sunday = 6
df['Dia_Semana'] = df['Fecha_Hora_Retiro'].dt.dayofweek
```

```python
# Encode categorical features


le_genero = LabelEncoder()
df['Genero_Usuario_encoded'] = le_genero.fit_transform(df['Genero_Usuario'])


le_estacion = LabelEncoder()
df['Estacion_Retiro_encoded'] = le_estacion.fit_transform(df['Ciclo_Estacion_Retiro'])
```
