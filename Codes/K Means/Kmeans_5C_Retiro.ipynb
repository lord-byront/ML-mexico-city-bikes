```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score
from datetime import datetime
```

```python
csv = 'archivo_ordenado.csv'
df = pd.read_csv(csv)
```

```python
# Ver las primeras filas del DataFrame
df.head()
```

```python
# Convertir las columnas de fecha y hora
df['Fecha_Retiro'] = pd.to_datetime(df['Fecha_Retiro'], format='%d/%m/%Y')
df['Hora_Retiro'] = pd.to_datetime(df['Hora_Retiro'], format='%H:%M:%S').dt.time
```

```python
# Crear nuevas columnas que representen la hora y el día
df['Hora_Num'] = df['Fecha_Retiro'].dt.hour + df['Hora_Retiro'].apply(lambda x: x.hour / 60)
df['Dia_Num'] = df['Fecha_Retiro'].dt.dayofweek  # Lunes=0, Domingo=6
```

```python
# Filtrar las columnas relevantes para el clustering
X = df[['latitud_Retiro', 'longitud_Retiro', 'Hora_Num', 'Dia_Num']]
```

```python
# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Gráfica de codo
inertia = []
k_values = range(1, 15)  # Cambia el rango según necesites

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotear la gráfica del codo

plt.plot(k_values, inertia, marker='o')
plt.title('Gráfica del Codo para K-Means')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.xticks(k_values)

plt.show()
```

```python
# Aplicar el algoritmo K-Means
n_clusters = 5  # Cambia el número de clusters según sea necesario
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_scaled)
```

```python
# Agregar las etiquetas de cluster al DataFrame
df['Cluster'] = kmeans.labels_
```

```python
# Visualización de los clusters
plt.scatter(df['longitud_Retiro'], df['latitud_Retiro'], c=df['Cluster'], cmap='viridis', alpha=0.5)
plt.title('Bicycle route clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Cluster')
plt.grid()
plt.show()
```

```python
# Análisis de los clusters
cluster_summary = df.groupby('Cluster').agg({
    'Genero_Usuario': 'count',
    'Edad_Usuario': 'mean',
    'Ciclo_Estacion_Retiro': 'first',
    'nombre_estacion_Retiro': 'first'
}).reset_index()

print(cluster_summary)
```

```python
# Evaluar rendimiento del modelo de clustering
inertia = kmeans.inertia_
print(f"Inercia: {inertia}")
```

```python
# Tomar una muestra más pequeña si es necesario
sample_size = 100000  # Ajusta según tus necesidades
X_sample = X_scaled[np.random.choice(X_scaled.shape[0], sample_size, replace=False)]

# Calcular el Silhouette Score
silhouette_avg = silhouette_score(X_sample, kmeans.labels_[:sample_size])
print(f"Silhouette Score: {silhouette_avg}")
```
